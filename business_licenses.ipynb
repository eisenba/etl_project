```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store CSV into a Dataframe 

```python
csv_file = "./Resources/licenses_data.csv"
licenses_data_df = pd.read_csv(csv_file)
licenses_data_df.head()
```

### Create new data with select columns and group 

```python
new_licenses_data_df = licenses_data_df[['LEGAL NAME','ZIP CODE', 'LICENSE DESCRIPTION', 'DATE ISSUED']].copy()
new_licenses_data_df.head()
```

```python
#groupby license type
grouped_licenses_df = new_licenses_data_df.groupby(['LICENSE DESCRIPTION', 'ZIP CODE', 'LEGAL NAME', 'DATE ISSUED'])
print(grouped_licenses_df)
grouped_licenses_df.count().head(40)
```
